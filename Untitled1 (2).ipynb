```python
from encodec import EncodecModel
from encodec.utils import convert_audio
import torchaudio
import numpy as np
import os
import torch

voice_dir = "assets/prompts/my_voice"
wav_path = os.path.join(voice_dir, "train_clean.wav")
npz_path = os.path.join(voice_dir, "train.npz")

model = EncodecModel.encodec_model_24khz()
model.set_target_bandwidth(6.0)
model.eval()

waveform, sr = torchaudio.load(wav_path)
waveform = convert_audio(waveform, sr, model.sample_rate, model.channels).unsqueeze(0)

with torch.no_grad():
    encoded_frames = model.encode(waveform)

codes = torch.cat([encoded[0] for encoded in encoded_frames], dim=-1).cpu().numpy()
np.savez_compressed(npz_path, fine_prompt=codes)

print("Done! Voice embedding saved at:", npz_path)
```

```python
import os
import numpy as np
from bark import generation
from bark.generation import CUR_PATH

def custom_load_history_prompt(history_prompt_input):
    history_prompt_input = os.path.join(*history_prompt_input.split("/"))
    path = os.path.join(CUR_PATH, "assets", "prompts", f"{history_prompt_input}.npz")
    if not os.path.exists(path):
        raise ValueError(f"Custom prompt file not found: {path}")
    return np.load(path, allow_pickle=True)

generation._load_history_prompt = custom_load_history_prompt
```

```python
import torch
import torchaudio
import numpy as np
from encodec.utils import convert_audio
from encodec.model import EncodecModel

input_path = "assets/prompts/my_voice/train_clean.wav"
output_path = "assets/prompts/my_voice/train.npz"

model = EncodecModel.encodec_model_24khz()
model.set_target_bandwidth(6.0)
model.eval()

waveform, sr = torchaudio.load(input_path)
waveform = convert_audio(waveform, sr, model.sample_rate, model.channels).unsqueeze(0)

with torch.no_grad():
    encoded_frames = model.encode(waveform)

codes = torch.cat([encoded[0] for encoded in encoded_frames], dim=-1).cpu().numpy()

np.savez(output_path,
         semantic_prompt=np.zeros((1, 1)),
         coarse_prompt=np.zeros((1, 1)),
         fine_prompt=codes)

print("Voice prompt saved to:", output_path)
```
